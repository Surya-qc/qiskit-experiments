```python
#Simulated Bell correlation measurement with histogram
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import numpy as np
import matplotlib.pyplot as plt
theta_1=float(input('Enter Theta 1 in degree: '))
theta_2=float(input('Enter Theta 2 in degree: '))
theta_1 = (np.pi/180)*theta_1
theta_2 = (np.pi/180)*theta_2
shots = 1000
qc=QuantumCircuit(2,2)
qc.h(0)
qc.cx(0,1)
qc.ry(-2*theta_1,0)
qc.ry(-2*theta_2,1)
qc.measure(0,0)
qc.measure(1,1)
simulator=AerSimulator()
counts=simulator.run(qc,shots=shots).result().get_counts()
N00=counts.get('00',0)
N01=counts.get('01',0)
N10=counts.get('10',0)
N11=counts.get('11',0)
E=(N00+N11-N01-N10)/shots
print(f'Value of E=  {E:.4f}')
print(f'Theoretical E = {np.cos(2*(theta_1 - theta_2)):.4f}')
fig = plot_histogram(counts, title="Simulated Bell correlation measurement")
fig.savefig("Simulated Bell correlation measurement.png", dpi=150, bbox_inches='tight')
display(fig)
```
